### Function to produce catalog based on dataset and description txt file

```python
def produce_catalog (dataset, description_txt_file):
  # Based on the dataset dataframe and the description of the columns in a txt file,
  # creates a catalog with field, data types and descriptions
  
  import pandas as pd

  columns = pd.DataFrame(dataset.columns, columns = ['campo'])
  columns.reset_index(drop=True, inplace=True)
  # columns

  data_types = pd.DataFrame(dataset.dtypes, columns = ['tipo de dado'])
  data_types.reset_index(drop=True, inplace=True)
  # data_types

  description = pd.read_csv(description_txt_file, header = None)
  description.reset_index(drop=True, inplace=True)
  description.columns = ['descrição']

  df_catalog = pd.concat ( [columns, data_types,description],axis = 1)
  return (df_catalog)
```
